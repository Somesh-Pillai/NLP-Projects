```python
corpus=[]
for i in range(len(sentences)):
  sents=re.sub('[^a-zA-Z]',' ',sentences[i])### replacing non characters to ' '
  sent_word_list=sents.lower().split()
  no_stop_words=[lemmatizer.lemmatize(w) for w in sent_word_list if w not in set(stopwords.words('english'))]
  sent_new=' '.join(no_stop_words)
  corpus.append(sent_new)
```

```python
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```python
nltk.download('state_union')
```

```python
train_text=state_union.raw("2005-GWBush.txt")
sample_text=state_union.raw("2006-GWBush.txt")

```

```python
custom_sent_tokenizer= PunktSentenceTokenizer(train_text)
tokenizer=custom_sent_tokenizer.tokenize(train_text)
```

CHunking makes simliar words together

**Regular Expression**

```python
def process_content():
  try:
    for i in tokenizer:
      words=nltk.word_tokenize(i)
      tags=nltk.pos_tag(words)
      chunkgram= r"""<Chunk: {<RB.?>*<VB.?>*<NNP>+<NN>?}"""
      chunks=nltk.RegexpParser(chunkgram)
      chunked= chunks.parse(tags)
      chunked.draw()

  except Exception as e:
    print(str(e))
```

```python
nltk.download('punkt')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
process_content()
```

Named Entity Recognition

```python
def process_nameE():
  try:
    for i in tokenizer:
      words=nltk.word_tokenize(i)
      tags=nltk.pos_tag(words)
      namedE=nltk.ne_chunk(tags, binary=True)
      namedE.draw()
  except Exception as e:
    print(str(e))
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
process_nameE()
```

### Word2Vec

Maintains the semantic meaning of the words

Each word represented as vector of 32 dimension

```python
from gensim.models import Word2Vec
```

```python
import re
text = re.sub(r'\[[0-9]*\]',' ',example)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
sentences=sent_tokenize(text)
```

```python
corpus=[]
for i in range(len(sentences)):
  words=word_tokenize(sentences[i])
  words=[word for word in words if word not in set(stopwords.words("english"))]
  corpus.append(words)
```

```python
corpus
```

```python
model = Word2Vec(corpus, min_count=1)

## skip words which occur only once
## depends on how big ur data is
```
